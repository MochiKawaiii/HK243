## Cơ Sở Lý Thuyết

### 1. Tìm Kiếm Mù
Tìm kiếm mù (Uninformed Search) là các thuật toán tìm kiếm không sử dụng thông tin heuristic, duyệt qua không gian trạng thái theo cách có hệ thống, dựa trên cấu trúc bài toán.

### 2. Tìm Kiếm theo Chiều Rộng (BFS)
- **Nguyên lý**: Khám phá tất cả các nút ở mức hiện tại trước khi chuyển sang mức tiếp theo.
- **Cấu trúc dữ liệu**: Hàng đợi (Queue).
- **Đặc điểm**:
  - Trong đồ thị không trọng số, BFS đảm bảo tìm đường đi ngắn nhất (theo số cạnh).
  - Trong đồ thị có trọng số, BFS không đảm bảo đường đi có tổng trọng số nhỏ nhất, vì nó không xem xét trọng số của các cạnh.

### 3. Tìm Kiếm theo Chiều Sâu (DFS)
- **Nguyên lý**: Khám phá một nhánh đến tận cùng trước khi quay lại thử nhánh khác.
- **Cấu trúc dữ liệu**: Ngăn xếp (Stack) hoặc đệ quy.
- **Đặc điểm**:
  - DFS không đảm bảo đường đi ngắn nhất trong cả đồ thị không trọng số và có trọng số.
  - Kết quả phụ thuộc vào thứ tự duyệt các nút kề.

### 4. Đồ Thị Không Trọng Số và Có Trọng Số
- **Đồ thị không trọng số**: Các cạnh có trọng số ngầm định bằng 1. BFS tối ưu theo số cạnh.
- **Đồ thị có trọng số**: Các cạnh có trọng số khác nhau (ví dụ: chi phí, khoảng cách). BFS và DFS không được thiết kế để tối ưu hóa tổng trọng số, dẫn đến kết quả không tối ưu.

### 5. Bài Toán 
Cho các đồ thị vô hướng (không trọng số và có trọng số) biểu diễn mạng lưới hoặc mê cung. Nhiệm vụ là tìm đường đi từ nút bắt đầu (**S**) đến nút đích (**G** hoặc **H**) bằng BFS và DFS, thông qua minh họa thủ công và lập trình với chú thích chi tiết.

### Phần 3: Triển Khai Mã Python

#### BFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 1**.

````python
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy BFS và in kết quả
path = bfs(graph1, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 1:", path)
````

#### DFS trên Đồ Thị Không Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 1**.

````python
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
graph1 = {
    'S': ['A', 'B', 'C'],
    'A': ['S', 'D'],
    'B': ['S', 'E'],
    'C': ['S'],
    'D': ['A', 'F'],
    'E': ['B', 'G'],
    'F': ['D'],
    'G': ['E']
}

# Chạy DFS và in kết quả
path = dfs(graph1, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 1:", path)
````

#### BFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng BFS trên **Đồ thị mẫu 5**.

````python
from collections import deque

def bfs_weighted(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu, đường đi, và tổng trọng số
    queue = deque([(start, [start], 0)])
    # Khởi tạo tập hợp các nút đã thăm
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút, đường đi, và tổng trọng số từ đầu hàng đợi
        (node, path, total_weight) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path, total_weight
        # Duyệt qua các nút kề và trọng số tương ứng
        for neighbor, weight in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề, đường đi mới, và tổng trọng số cập nhật vào hàng đợi
                queue.append((neighbor, path + [neighbor], total_weight + weight))
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy BFS và in kết quả
path, weight = bfs_weighted(graph5, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

#### DFS trên Đồ Thị Có Trọng Số
Chạy ô mã dưới đây để kiểm chứng DFS trên **Đồ thị mẫu 5**.

````python
def dfs_weighted(graph, start, goal, visited=None, path=None, total_weight=0):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path, total_weight
    
    # Duyệt qua các nút kề và trọng số tương ứng
    for neighbor, weight in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi và tổng trọng số mới
            new_path, new_weight = dfs_weighted(graph, neighbor, goal, visited, path + [neighbor], total_weight + weight)
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path, new_weight
    # Trả về None và 0 nếu không tìm thấy đường đi
    return None, 0

# Định nghĩa Đồ thị mẫu 5 (có trọng số)
graph5 = {
    'S': [('A', 1), ('B', 2), ('C', 3)],
    'A': [('S', 1), ('D', 4), ('G', 3)],
    'B': [('S', 2), ('G', 5)],
    'C': [('S', 3)],
    'D': [('A', 4), ('G', 6)],
    'G': [('B', 5), ('D', 6), ('A', 3)]
}

# Chạy DFS và in kết quả
path, weight = dfs_weighted(graph5, 'S', 'G')
print("Đường đi DFS trên Đồ thị mẫu 5:", path, "Tổng trọng số:", weight)
````

````python
# Minh hoạt thủ công ở đây: 
#### Đồ Thị Mẫu 2: Có Chu Trình, Không Trọng Số
##### Mermaid Diagram
'''mermaid
graph TD
    S --> A
    S --> B
    A --> B
    A --> C
    B --> D
    C --> D
    D --> G

- **Các cạnh**: S-A, S-B, A-B, A-C, B-D, C-D, D-G.
'''
'''
Minh họa BFS
1. Khởi tạo: Hàng đợi = [(S, [S])], Đã thăm = {S}
2. Lấy S, Thêm A, B: Hàng đợi = [(A, [S, A]), (B, [S, B])], Đã thăm = {S, A, B}
3. Lấy A, Thêm C(B đã thăm): Hàng đợi = [(B, [S, B])], [(C, [S, A, C])], Đã thăm = {S, A, B, C}
4. Lấy B, Thêm D: Hàng đợi = [(C, [S, A, C])], [(D, [S, B, D])], Đã thăm = {S, A, B, C, D}
5. Lấy C, D đã thăm: Hàng đợi = [(D, [S, B, D])], Đã thăm = {S, A, B, C, D}
6. Lấy D, Thêm G: Hàng đợi = [(G, [S, B, D, G])], Đã thăm = {S, A, B, C, D, G}
7. Lấy G làm đích trả về đường đi: [S, B, D, G] 
Kết quả: Đường đi BFS: S -> B -> D -> G (độ dài 3, ngắn nhất)

Minh họa DFS
1. Khởi tạo: Ngăn xếp = [(S, [S])], Đã thăm = {S}
2. Lấy S, Thêm A: Ngăn xếp = [(A, [S, A])], Đã thăm = {S, A}
3. Lấy A, Thêm B: Ngắn xếp = [(B, [S, A, B])], Đã thăm = {S, A, B}
4. Lấy B, Thêm D: Ngăn xếp = [(D, [S, A, B, D])], Đã thăm = {S, A, B, D}
5. Lấy D, Thêm G: Ngăn xếp = [(G, [S, A, B, D, G])], Đã thăm = {S, A, B, D, G}
6. Lấy G làm đích trả về đường đi: [S, A, B, D, G]
Kết quả: Đường đi DFS: S-> A -> B-> D-> G (đào sâu trước, không đảm bảo ngắn nhất)
'''
#### Đồ Thị Mẫu 3: Không Có Đường Đi Tới Đích, Không Trọng Số
##### Mermaid Diagram
'''mermaid
graph TD
    S --> A
    S --> B
    A --> C
    B --> D

- **Các cạnh**: S-A, S-B, A-C, B-D.
'''
'''
Minh họa BFS
1. Khởi tạo: Hàng đợi = [(S,[S])], Đã thăm = {S}
2. Lấy S, Thêm A, B → Hàng đợi = [(A,[S,A]), (B,[S,B])], Đã thăm = {S,A,B}
3. Lấy A, Thêm C → Hàng đợi = [(B,[S,B]), (C,[S,A,C])], Đã thăm = {S,A,B,C}
4. Lấy B, Thêm D → Hàng đợi = [(C,[S,A,C]), (D,[S,B,D])], Đã thăm = {S,A,B,C,D}
5. Lấy C, (D đã thăm) → Hàng đợi = [(D,[S,B,D])], Đã thăm = {S,A,B,C,D}
6. Lấy D, không có láng giềng mới → Hàng đợi = [] (rỗng)
7. Hàng đợi rỗng ⇒ không tìm thấy G
Kết quả BFS: G không nằm trong thành phần liên thông của S

Minh họa DFS
1. Khởi tạo: Ngăn xếp = [(S,[S])], Đã thăm = {S}
2. Lấy S, Thêm B, A (duyệt ngược alphabet để A được lấy trước)
→ Ngăn xếp = [(B,[S,B]), (A,[S,A])], Đã thăm = {S,A,B}
3. Lấy A, Thêm C → Ngăn xếp = [(B,[S,B]), (C,[S,A,C])], Đã thăm = {S,A,B,C}
4. Lấy C, không có láng giềng mới → Ngăn xếp = [(B,[S,B])]
5. Lấy B, Thêm D → Ngăn xếp = [(D,[S,B,D])], Đã thăm = {S,A,B,C,D}
6. Lấy D, không có láng giềng mới → Ngăn xếp = [] (rỗng)
7. Ngăn xếp rỗng ⇒ không tìm thấy G

Kết quả DFS: Cũng không tìm thấy G.
'''
````

## Bài Tập lập trình

#### Bài Tập Lập Trình: BFS trên Đồ Thị Mẫu 2
Chạy BFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
from collections import deque

def bfs(graph, start, goal):
    # Khởi tạo hàng đợi với nút bắt đầu và đường đi ban đầu
    queue = deque([(start, [start])])
    # Khởi tạo tập hợp các nút đã thăm để tránh lặp
    visited = set([start])
    
    # Lặp cho đến khi hàng đợi rỗng
    while queue:
        # Lấy nút đầu tiên và đường đi tương ứng từ hàng đợi
        (node, path) = queue.popleft()
        # Kiểm tra nếu nút hiện tại là đích
        if node == goal:
            return path
        # Duyệt qua các nút kề của nút hiện tại
        for neighbor in graph[node]:
            # Nếu nút kề chưa được thăm
            if neighbor not in visited:
                # Thêm nút kề vào tập đã thăm
                visited.add(neighbor)
                # Thêm nút kề và đường đi mới vào hàng đợi
                queue.append((neighbor, path + [neighbor]))
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 2 (có chu trình, không trọng số)
# Dựa trên mermaid diagram: S->A, S->B, A->B, A->C, B->D, C->D, D->G
graph2 = {
    'S': ['A', 'B'],        
    'A': ['S', 'B', 'C'],   
    'B': ['S', 'A', 'D'],  
    'C': ['A', 'D'],        
    'D': ['B', 'C', 'G'],   
    'G': ['D']             
}

# Chạy BFS trên Đồ thị mẫu 2
path2 = bfs(graph2, 'S', 'G')
print("Đường đi BFS trên Đồ thị mẫu 2:", path2)

# Chú thích chi tiết về đồ thị mẫu 2:
print("\n=== PHÂN TÍCH ĐỒ THỊ MẪU 2 ===")
print("- Đồ thị có chu trình: S -> A -> B -> S")
print("- BFS sẽ tìm đường đi ngắn nhất từ S đến G")
print("- Các đường đi có thể: S->B->D->G hoặc S->A->C->D->G")
print("- BFS sẽ chọn đường đi có ít bước nhất")
print("- Thuật toán tránh lặp vô hạn nhờ tập visited")
````

#### Bài Tập Lập Trình: DFS trên Đồ Thị Mẫu 2
Chạy DFS trên **Đồ thị mẫu 2** với chú thích chi tiết.

````python
# code here
def dfs(graph, start, goal, visited=None, path=None):
    # Khởi tạo tập đã thăm và đường đi nếu chưa có
    if visited is None:
        visited = set()
    if path is None:
        path = [start]
    
    # Thêm nút hiện tại vào tập đã thăm
    visited.add(start)
    # Kiểm tra nếu nút hiện tại là đích
    if start == goal:
        return path
    
    # Duyệt qua các nút kề của nút hiện tại
    for neighbor in graph[start]:
        # Nếu nút kề chưa được thăm
        if neighbor not in visited:
            # Gọi đệ quy DFS trên nút kề với đường đi mới
            new_path = dfs(graph, neighbor, goal, visited, path + [neighbor])
            # Nếu tìm thấy đường đi, trả về
            if new_path:
                return new_path
    # Trả về None nếu không tìm thấy đường đi
    return None

# Định nghĩa Đồ thị mẫu 1 (không trọng số)
# Định nghĩa Đồ thị mẫu 2 (có chu trình, không trọng số)
graph2 = {
    'S': ['A', 'B'],      # S kết nối với A và B
    'A': ['S', 'B', 'C'], # A kết nối với S, B, C (tạo chu trình S-A-B-S)
    'B': ['S', 'A', 'D'], # B kết nối với S, A, D (tạo chu trình S-A-B-S)
    'C': ['A', 'D'],      # C kết nối với A và D
    'D': ['B', 'C', 'G'], # D kết nối với B, C, G
    'G': ['D']            # G chỉ kết nối với D (đích cuối)
}

# Chạy DFS trên đồ thị mẫu 2 và in kết quả
print("\n--- BÀI MẪU 2: ĐỒ THỊ CÓ CHU TRÌNH ---")
print("Đồ thị:", graph2)
path2 = dfs(graph2, 'S', 'G')
print("Đường đi DFS từ S đến G:", path2)

# Giải thích chi tiết quá trình DFS:
print("\n--- GIẢI THÍCH CHI TIẾT QUÁ TRÌNH DFS ---")
print("1. Bắt đầu từ S, visited = {S}")
print("2. Từ S có thể đi đến: A, B")
print("3. Chọn A trước (theo thứ tự trong danh sách)")
print("4. Từ A có thể đi đến: S(đã thăm), B, C")
print("5. Chọn B tiếp theo, visited = {S, A, B}")
print("6. Từ B có thể đi đến: S(đã thăm), A(đã thăm), D")
print("7. Chọn D, visited = {S, A, B, D}")
print("8. Từ D có thể đi đến: B(đã thăm), C, G")
print("9. Chọn C, visited = {S, A, B, D, C}")
print("10. Từ C không có nút mới để thăm, quay lại D")
print("11. Từ D chọn G, tìm thấy đích!")
print("12. Đường đi cuối cùng: S → A → B → D → G")
````

### Sửa mã BFS để đếm số nút đã thăm trên **Đồ thị mẫu 4**. In ra kết quả và đảm bảo mã có chú thích đầy đủ.

````python
# code here
from collections import deque

def bfs_count_visited(graph, start):
    """
    Thực hiện thuật toán BFS và đếm số nút đã thăm
    
    Args:
        graph (dict): Đồ thị dưới dạng danh sách kề
        start (str): Nút bắt đầu
    
    Returns:
        tuple: (số nút đã thăm, danh sách nút đã thăm theo thứ tự)
    """
    visited = set()  # Tập hợp các nút đã thăm
    queue = deque([start])  # Hàng đợi cho BFS
    visit_order = []  # Danh sách lưu thứ tự thăm các nút
    
    visited.add(start)
    visit_order.append(start)
    
    while queue:
        current = queue.popleft()  # Lấy nút đầu tiên trong hàng đợi
        
        # Duyệt tất cả các nút kề
        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                visit_order.append(neighbor)
                queue.append(neighbor)
    
    return len(visited), visit_order

# Định nghĩa đồ thị mẫu 4 dưới dạng danh sách kề
graph_sample_4 = {
    'S': ['A', 'C'],
    'A': ['S', 'B', 'D'],
    'B': ['A', 'E'],
    'C': ['S', 'D', 'F'],
    'D': ['A', 'C', 'E'],
    'E': ['B', 'D', 'G'],
    'F': ['C', 'G'],
    'G': ['E', 'F']
}

# Thực hiện BFS từ nút 'S'
start_node = 'S'
visited_count, visit_sequence = bfs_count_visited(graph_sample_4, start_node)

# In kết quả
print("=== KẾT QUẢ BFS TRÊN ĐỒ THỊ MẪU 4 ===")
print(f"Nút bắt đầu: {start_node}")
print(f"Số nút đã thăm: {visited_count}")
print(f"Thứ tự thăm các nút: {' -> '.join(visit_sequence)}")
print(f"Danh sách các nút đã thăm: {visit_sequence}")

# Kiểm tra xem có thăm hết tất cả các nút không
total_nodes = len(graph_sample_4)
print(f"\nTổng số nút trong đồ thị: {total_nodes}")
if visited_count == total_nodes:
    print("✓ Đã thăm hết tất cả các nút trong đồ thị")
else:
    print(f"✗ Chỉ thăm được {visited_count}/{total_nodes} nút")
````

````python
# code here
# 1. Viết mã Python để chạy BFS và DFS trên **Đồ thị mẫu 6** và **Đồ thị mẫu 7**. Định nghĩa đồ thị dưới dạng từ điển và thêm chú thích chi tiết.
from collections import deque, defaultdict
import heapq

# Định nghĩa đồ thị mẫu 6 (có trọng số)
graph_6_weighted = {
    'S': [('A', 2), ('C', 5)],
    'A': [('B', 3), ('D', 4)],
    'B': [('E', 6)],
    'C': [('D', 7), ('F', 9)],
    'D': [('E', 8)],
    'E': [('H', 10)],
    'F': [('G', 12)],
    'G': [('H', 15)],
    'H': []
}

# Định nghĩa đồ thị mẫu 6 (không trọng số cho BFS/DFS)
graph_6_unweighted = {
    'S': ['A', 'C'],
    'A': ['B', 'D'],
    'B': ['E'],
    'C': ['D', 'F'],
    'D': ['E'],
    'E': ['H'],
    'F': ['G'],
    'G': ['H'],
    'H': []
}

# Định nghĩa đồ thị mẫu 7 (không trọng số)
graph_7 = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': []
}

def bfs_shortest_path(graph, start, goal):
    """
    Thuật toán BFS tìm đường đi ngắn nhất (theo số bước)
    Args:
        graph: Đồ thị dạng từ điển
        start: Đỉnh bắt đầu
        goal: Đỉnh kết thúc
    Returns:
        Tuple (đường đi, độ dài đường đi)
    """
    if start == goal:
        return [start], 0
    
    queue = deque([(start, [start])])  # (đỉnh hiện tại, đường đi)
    visited = {start}
    
    while queue:
        current, path = queue.popleft()
        
        # Duyệt các đỉnh kề
        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                new_path = path + [neighbor]
                if neighbor == goal:
                    return new_path, len(new_path) - 1
                
                visited.add(neighbor)
                queue.append((neighbor, new_path))
    
    return None, float('inf')  # Không tìm thấy đường đi

def dfs_all_paths(graph, start, goal, path=None):
    """
    Thuật toán DFS tìm tất cả đường đi từ start đến goal
    Args:
        graph: Đồ thị dạng từ điển
        start: Đỉnh bắt đầu
        goal: Đỉnh kết thúc
        path: Đường đi hiện tại
    Returns:
        List các đường đi
    """
    if path is None:
        path = []
    
    path = path + [start]
    
    if start == goal:
        return [path]
    
    paths = []
    for neighbor in graph.get(start, []):
        if neighbor not in path:  # Tránh chu trình
            new_paths = dfs_all_paths(graph, neighbor, goal, path)
            paths.extend(new_paths)
    
    return paths

def dijkstra_shortest_path(graph, start, goal):
    """
    Thuật toán Dijkstra tìm đường đi ngắn nhất theo trọng số
    Args:
        graph: Đồ thị có trọng số dạng từ điển
        start: Đỉnh bắt đầu
        goal: Đỉnh kết thúc
    Returns:
        Tuple (đường đi, tổng trọng số)
    """
    # Khởi tạo khoảng cách và đường đi
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    previous = {}
    
    # Priority queue: (khoảng cách, đỉnh)
    pq = [(0, start)]
    visited = set()
    
    while pq:
        current_dist, current = heapq.heappop(pq)
        
        if current in visited:
            continue
            
        visited.add(current)
        
        if current == goal:
            # Xây dựng đường đi
            path = []
            while current is not None:
                path.append(current)
                current = previous.get(current)
            return path[::-1], distances[goal]
        
        # Duyệt các đỉnh kề
        for neighbor, weight in graph.get(current, []):
            if neighbor not in visited:
                new_dist = current_dist + weight
                if new_dist < distances[neighbor]:
                    distances[neighbor] = new_dist
                    previous[neighbor] = current
                    heapq.heappush(pq, (new_dist, neighbor))
    
    return None, float('inf')

# Chạy thuật toán trên đồ thị mẫu 6
print("=== ĐỒ THỊ MẪU 6 ===")
print("\n1. BFS (đường đi ngắn nhất theo số bước):")
path_bfs_6, length_bfs_6 = bfs_shortest_path(graph_6_unweighted, 'S', 'H')
print(f"Đường đi: {' -> '.join(path_bfs_6)}")
print(f"Số bước: {length_bfs_6}")

print("\n2. DFS (tất cả đường đi):")
all_paths_6 = dfs_all_paths(graph_6_unweighted, 'S', 'H')
for i, path in enumerate(all_paths_6, 1):
    print(f"Đường đi {i}: {' -> '.join(path)} (Số bước: {len(path)-1})")

print("\n3. Dijkstra (đường đi ngắn nhất theo trọng số):")
path_dijkstra_6, weight_dijkstra_6 = dijkstra_shortest_path(graph_6_weighted, 'S', 'H')
print(f"Đường đi: {' -> '.join(path_dijkstra_6)}")
print(f"Tổng trọng số: {weight_dijkstra_6}")

# Chạy thuật toán trên đồ thị mẫu 7
print("\n\n=== ĐỒ THỊ MẪU 7 ===")
print("\n1. BFS (đường đi ngắn nhất theo số bước):")
path_bfs_7, length_bfs_7 = bfs_shortest_path(graph_7, 'S', 'H')
print(f"Đường đi: {' -> '.join(path_bfs_7)}")
print(f"Số bước: {length_bfs_7}")

print("\n2. DFS (tất cả đường đi - giới hạn 5 đường đầu tiên):")
all_paths_7 = dfs_all_paths(graph_7, 'S', 'H')
for i, path in enumerate(all_paths_7[:5], 1):  # Chỉ hiển thị 5 đường đầu
    print(f"Đường đi {i}: {' -> '.join(path)} (Số bước: {len(path)-1})")
print(f"Tổng số đường đi tìm được: {len(all_paths_7)}")

# So sánh kết quả
print("\n\n=== SO SÁNH KẾT QUẢ ===")
print("Đồ thị mẫu 6:")
print(f"- BFS (số bước): {length_bfs_6}")
print(f"- Dijkstra (trọng số): {weight_dijkstra_6}")
print(f"- Số đường đi khác nhau: {len(all_paths_6)}")

print("\nĐồ thị mẫu 7:")
print(f"- BFS (số bước): {length_bfs_7}")
print(f"- Số đường đi khác nhau: {len(all_paths_7)}")
````

````python
# code here
# 2. Sửa mã BFS để lưu và in tất cả các đường đi từ S đến H trên **Đồ thị mẫu 7** (không trọng số). Đảm bảo mã có chú thích đầy đủ.
from collections import deque

def bfs_all_paths(graph, start, end):
    """
    Tìm tất cả các đường đi từ start đến end bằng BFS
    
    Args:
        graph: Dictionary biểu diễn đồ thị (adjacency list)
        start: Đỉnh bắt đầu
        end: Đỉnh kết thúc
    
    Returns:
        List chứa tất cả các đường đi từ start đến end
    """
    # Queue để lưu các đường đi hiện tại (mỗi phần tử là một đường đi)
    queue = deque([[start]])
    all_paths = []  # Danh sách lưu tất cả các đường đi hoàn chỉnh
    
    while queue:
        # Lấy đường đi hiện tại từ queue
        current_path = queue.popleft()
        current_node = current_path[-1]  # Đỉnh cuối cùng của đường đi hiện tại
        
        # Nếu đã đến đích, thêm đường đi vào kết quả
        if current_node == end:
            all_paths.append(current_path)
            continue
        
        # Duyệt tất cả các đỉnh kế tiếp
        for neighbor in graph.get(current_node, []):
            # Tránh chu trình: chỉ thêm đỉnh chưa có trong đường đi hiện tại
            if neighbor not in current_path:
                # Tạo đường đi mới bằng cách thêm đỉnh kế tiếp
                new_path = current_path + [neighbor]
                queue.append(new_path)
    
    return all_paths

def print_all_paths(paths, start, end):
    """
    In tất cả các đường đi từ start đến end
    
    Args:
        paths: Danh sách các đường đi
        start: Đỉnh bắt đầu
        end: Đỉnh kết thúc
    """
    if not paths:
        print(f"Không có đường đi từ {start} đến {end}")
        return
    
    print(f"Tất cả các đường đi từ {start} đến {end}:")
    for i, path in enumerate(paths, 1):
        print(f"Đường đi {i}: {' -> '.join(path)} (độ dài: {len(path)})")
    
    print(f"\nTổng số đường đi: {len(paths)}")
    
    # Tìm đường đi ngắn nhất
    shortest_length = min(len(path) for path in paths)
    shortest_paths = [path for path in paths if len(path) == shortest_length]
    
    print(f"Đường đi ngắn nhất (độ dài {shortest_length}):")
    for path in shortest_paths:
        print(f"  {' -> '.join(path)}")

# Định nghĩa đồ thị mẫu 7 (không trọng số, mật độ cao)
graph_7 = {
    'S': ['A', 'D', 'E'],           # S kết nối với A, D, E
    'A': ['B', 'D', 'E'],           # A kết nối với B, D, E
    'B': ['C', 'E', 'F'],           # B kết nối với C, E, F
    'C': ['F', 'G'],                # C kết nối với F, G
    'D': ['E'],                     # D kết nối với E
    'E': ['F', 'H'],                # E kết nối với F, H
    'F': ['G', 'H'],                # F kết nối với G, H
    'G': ['H'],                     # G kết nối với H
    'H': []                         # H là đỉnh đích
}

# Chương trình chính
if __name__ == "__main__":
    # Tìm tất cả đường đi từ S đến H
    start_node = 'S'
    end_node = 'H'
    
    print(f"\n=== BFS TÌM TẤT CẢ ĐƯỜNG ĐI TỪ {start_node} ĐẾN {end_node} ===")
    all_paths = bfs_all_paths(graph_7, start_node, end_node)
    print_all_paths(all_paths, start_node, end_node)
    
    # Phân tích thêm
    if all_paths:
        path_lengths = [len(path) for path in all_paths]
        print(f"\nPhân tích:")
        print(f"  - Độ dài ngắn nhất: {min(path_lengths)}")
        print(f"  - Độ dài dài nhất: {max(path_lengths)}")
````

````python
# code here
# 3. Thêm một cạnh mới vào **Đồ thị mẫu 6** (ví dụ: B-H với trọng số 20). Chạy lại BFS và DFS, phân tích sự thay đổi trong đường đi và tổng trọng số.
from collections import deque, defaultdict
import heapq

class WeightedGraph:
    def __init__(self):
        self.graph = defaultdict(list)
    
    def add_edge(self, u, v, weight):
        self.graph[u].append((v, weight))
    
    def bfs_shortest_path(self, start, end):
        """BFS để tìm đường đi với ít cạnh nhất (không xét trọng số)"""
        queue = deque([(start, [start], 0)])
        visited = set()
        
        while queue:
            node, path, total_weight = queue.popleft()
            
            if node in visited:
                continue
            visited.add(node)
            
            if node == end:
                return path, total_weight
            
            for neighbor, weight in self.graph[node]:
                if neighbor not in visited:
                    queue.append((neighbor, path + [neighbor], total_weight + weight))
        
        return None, float('inf')
    
    def dfs_all_paths(self, start, end, visited=None, path=None, weight=0):
        """DFS để tìm tất cả đường đi"""
        if visited is None:
            visited = set()
        if path is None:
            path = []
        
        visited.add(start)
        path.append(start)
        
        if start == end:
            yield path.copy(), weight
        else:
            for neighbor, edge_weight in self.graph[start]:
                if neighbor not in visited:
                    yield from self.dfs_all_paths(neighbor, end, visited, path, weight + edge_weight)
        
        path.pop()
        visited.remove(start)
    
    def dijkstra_shortest_path(self, start, end):
        """Dijkstra để tìm đường đi ngắn nhất theo trọng số"""
        distances = defaultdict(lambda: float('inf'))
        distances[start] = 0
        previous = {}
        pq = [(0, start)]
        visited = set()
        
        while pq:
            current_dist, current = heapq.heappop(pq)
            
            if current in visited:
                continue
            visited.add(current)
            
            if current == end:
                # Xây dựng đường đi
                path = []
                while current in previous:
                    path.append(current)
                    current = previous[current]
                path.append(start)
                return path[::-1], distances[end]
            
            for neighbor, weight in self.graph[current]:
                distance = current_dist + weight
                
                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    previous[neighbor] = current
                    heapq.heappush(pq, (distance, neighbor))
        
        return None, float('inf')

def analyze_graph():
    print("=== PHÂN TÍCH ĐỒ THỊ MẪU 6 ===\n")
    
    # Tạo đồ thị gốc
    graph_original = WeightedGraph()
    edges_original = [
        ('S', 'A', 2), ('S', 'C', 5), ('A', 'B', 3), ('A', 'D', 4),
        ('B', 'E', 6), ('C', 'D', 7), ('C', 'F', 9), ('D', 'E', 8),
        ('E', 'H', 10), ('F', 'G', 12), ('G', 'H', 15)
    ]
    
    for u, v, w in edges_original:
        graph_original.add_edge(u, v, w)
    
    # Tạo đồ thị mới với cạnh B-H (trọng số 20)
    graph_new = WeightedGraph()
    edges_new = edges_original + [('B', 'H', 20)]
    
    for u, v, w in edges_new:
        graph_new.add_edge(u, v, w)
    
    def analyze_single_graph(graph, title):
        print(f"--- {title} ---")
        
        # BFS
        bfs_path, bfs_weight = graph.bfs_shortest_path('S', 'H')
        print(f"BFS (ít cạnh nhất): {' -> '.join(bfs_path)}")
        print(f"Tổng trọng số BFS: {bfs_weight}")
        
        # Dijkstra (đường đi ngắn nhất theo trọng số)
        dijkstra_path, dijkstra_weight = graph.dijkstra_shortest_path('S', 'H')
        print(f"Dijkstra (ngắn nhất): {' -> '.join(dijkstra_path)}")
        print(f"Tổng trọng số Dijkstra: {dijkstra_weight}")
        
        # DFS - tìm tất cả đường đi và chọn ngắn nhất
        all_paths = list(graph.dfs_all_paths('S', 'H'))
        if all_paths:
            shortest_dfs = min(all_paths, key=lambda x: x[1])
            print(f"DFS (ngắn nhất trong tất cả): {' -> '.join(shortest_dfs[0])}")
            print(f"Tổng trọng số DFS: {shortest_dfs[1]}")
            print(f"Số đường đi tìm được: {len(all_paths)}")
        
        print()
    
    # Phân tích đồ thị gốc
    analyze_single_graph(graph_original, "ĐỒ THỊ GỐC")
    
    # Phân tích đồ thị mới
    analyze_single_graph(graph_new, "ĐỒ THỊ SAU KHI THÊM CẠNH B-H (20)")
    
    # So sánh
    print("--- SO SÁNH ---")
    
    # Đường đi ngắn nhất gốc
    _, orig_weight = graph_original.dijkstra_shortest_path('S', 'H')
    new_path, new_weight = graph_new.dijkstra_shortest_path('S', 'H')
    
    print(f"Đường đi ngắn nhất gốc: trọng số {orig_weight}")
    print(f"Đường đi ngắn nhất mới: {' -> '.join(new_path)}, trọng số {new_weight}")
    
    if new_weight < orig_weight:
        print(f"✓ Cải thiện {orig_weight - new_weight} đơn vị")
    elif new_weight > orig_weight:
        print(f"✗ Tăng {new_weight - orig_weight} đơn vị")
    else:
        print("= Không thay đổi")
    
    # Phân tích đường đi mới qua B-H
    print(f"\nĐường đi mới qua B-H:")
    path_through_b = ['S', 'A', 'B', 'H']
    weight_through_b = 2 + 3 + 20  # S-A + A-B + B-H
    print(f"{' -> '.join(path_through_b)}: {weight_through_b}")
    print(f"So với đường ngắn nhất: {weight_through_b - new_weight:+} đơn vị")

if __name__ == "__main__":
    analyze_graph()
````

````python
# code here
# 1. Đo thời gian chạy của BFS và DFS trên **Đồ thị mẫu 6** và **Đồ thị mẫu 7** bằng thư viện `time`. So sánh hiệu suất và thêm chú thích vào mã.
from __future__ import annotations
import time
import statistics
from collections import deque
from typing import Dict, List, Tuple, Optional, Set

# ──────────────────────────────────────────────────────────────────────────────
# ĐỒ THỊ MẪU
# ──────────────────────────────────────────────────────────────────────────────
GraphWeighted = Dict[str, List[Tuple[str, int]]]
GraphUnweighted = Dict[str, List[str]]

graph6: GraphWeighted = {
    'S': [('A', 2), ('C', 5)],
    'A': [('B', 3), ('D', 4)],
    'B': [('E', 6)],
    'C': [('D', 7), ('F', 9)],
    'D': [('E', 8)],
    'E': [('H', 10)],
    'F': [('G', 12)],
    'G': [('H', 15)],
    'H': []
}

graph7: GraphUnweighted = {
    'S': ['A', 'D', 'E'],
    'A': ['B', 'D', 'E'],
    'B': ['C', 'E', 'F'],
    'C': ['F', 'G'],
    'D': ['E'],
    'E': ['F', 'H'],
    'F': ['G', 'H'],
    'G': ['H'],
    'H': []
}

# ──────────────────────────────────────────────────────────────────────────────
# THUẬT TOÁN TÌM 1 ĐƯỜNG (KHÔNG TỐI ƯU TRỌNG SỐ)
# ──────────────────────────────────────────────────────────────────────────────
def bfs_weighted(graph: GraphWeighted, start: str, target: str) -> Optional[List[str]]:
    """BFS trên đồ thị CÓ trọng số (coi mọi cạnh = 1, chỉ cần tìm 1 đường)."""
    queue: deque[List[str]] = deque([[start]])
    visited: Set[str] = {start}

    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == target:
            return path
        for neighbor, _ in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])
    return None


def dfs_weighted(graph: GraphWeighted, start: str, target: str,
                 visited: Set[str] | None = None,
                 path: List[str] | None = None) -> Optional[List[str]]:
    """DFS trên đồ thị CÓ trọng số – không sao chép set/list mỗi khung lệnh."""
    if visited is None:
        visited = set()
    if path is None:
        path = []
    visited.add(start)
    path.append(start)

    if start == target:
        return path.copy()

    for neighbor, _ in graph[start]:
        if neighbor not in visited:
            res = dfs_weighted(graph, neighbor, target, visited, path)
            if res:
                return res
    path.pop()
    visited.remove(start)
    return None


def bfs_unweighted(graph: GraphUnweighted, start: str, target: str) -> Optional[List[str]]:
    queue: deque[List[str]] = deque([[start]])
    visited: Set[str] = {start}

    while queue:
        path = queue.popleft()
        node = path[-1]
        if node == target:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(path + [neighbor])
    return None


def dfs_unweighted(graph: GraphUnweighted, start: str, target: str,
                   visited: Set[str] | None = None,
                   path: List[str] | None = None) -> Optional[List[str]]:
    if visited is None:
        visited = set()
    if path is None:
        path = []
    visited.add(start)
    path.append(start)

    if start == target:
        return path.copy()

    for neighbor in graph[start]:
        if neighbor not in visited:
            res = dfs_unweighted(graph, neighbor, target, visited, path)
            if res:
                return res
    path.pop()
    visited.remove(start)
    return None

# ──────────────────────────────────────────────────────────────────────────────
# HÀM BENCHMARK
# ──────────────────────────────────────────────────────────────────────────────
def benchmark(fn, *args, repeats: int = 10_000):
    """
    Chạy `fn(*args)` `repeats` lần, trả về (kết_quả_đầu_tiên, thời_gian_trung_bình).
    Số lặp mặc định 10 000 để giảm nhiễu (có thể giảm cho máy yếu).
    """
    runtimes: List[float] = []
    result = None
    for i in range(repeats):
        start = time.perf_counter()
        result = fn(*args)
        runtimes.append(time.perf_counter() - start)
        if i == 0:
            first_result = result  # giữ lại kết quả đầu để in
    return first_result, statistics.mean(runtimes)

# ──────────────────────────────────────────────────────────────────────────────
# ĐO HIỆU NĂNG
# ──────────────────────────────────────────────────────────────────────────────
def measure_performance():
    print("=== ĐỒ THỊ 6 (CÓ TRỌNG SỐ, NHIỀU ĐƯỜNG) ===")
    bfs_res6, bfs_time6 = benchmark(bfs_weighted, graph6, 'S', 'H')
    dfs_res6, dfs_time6 = benchmark(dfs_weighted, graph6, 'S', 'H')
    ratio6 = "N/A" if bfs_time6 == 0 else f"{dfs_time6 / bfs_time6:.2f}"

    print(f"BFS: {bfs_res6}")
    print(f"Thời gian BFS (trung bình): {bfs_time6:.8e} s")
    print(f"DFS: {dfs_res6}")
    print(f"Thời gian DFS (trung bình): {dfs_time6:.8e} s")
    print(f"Tỉ lệ DFS/BFS: {ratio6}")

    print("\n=== ĐỒ THỊ 7 (KHÔNG TRỌNG SỐ, MẬT ĐỘ CAO) ===")
    bfs_res7, bfs_time7 = benchmark(bfs_unweighted, graph7, 'S', 'H')
    dfs_res7, dfs_time7 = benchmark(dfs_unweighted, graph7, 'S', 'H')
    ratio7 = "N/A" if bfs_time7 == 0 else f"{dfs_time7 / bfs_time7:.2f}"

    print(f"BFS: {bfs_res7}")
    print(f"Thời gian BFS (trung bình): {bfs_time7:.8e} s")
    print(f"DFS: {dfs_res7}")
    print(f"Thời gian DFS (trung bình): {dfs_time7:.8e} s")
    print(f"Tỉ lệ DFS/BFS: {ratio7}")

    print("\n=== NHẬN XÉT NHANH ===")
    print("• BFS duyệt theo lớp, ít bị lạc vào nhánh sâu → thường nhanh, đặc biệt đồ thị rậm.")
    print("• DFS dùng back-tracking tiết kiệm bộ nhớ hơn copy-path nhưng vẫn có thể chậm nếu nhánh sâu.")
    print("• Trên đồ thị nhỏ, cả hai thuật toán đều rất nhanh; cần lặp nhiều lần mới đo được chênh lệch.")


if __name__ == "__main__":
    measure_performance()
````

````python
# code here
# 2. Thiết kế một đồ thị có trọng số với ít nhất 10 nút và 15 cạnh. Chạy BFS và DFS, phân tích đường đi và tổng trọng số. Đảm bảo mã có chú thích chi tiết.
from collections import defaultdict, deque
import heapq

class WeightedGraph:
    def __init__(self):
        """
        Khởi tạo đồ thị có trọng số
        - graph: dictionary chứa danh sách kề với trọng số
        - vertices: tập hợp các đỉnh
        """
        self.graph = defaultdict(list)
        self.vertices = set()
    
    def add_edge(self, u, v, weight):
        """
        Thêm cạnh có trọng số vào đồ thị (vô hướng)
        """
        self.graph[u].append((v, weight))
        self.graph[v].append((u, weight))
        self.vertices.add(u)
        self.vertices.add(v)
    
    def bfs(self, start, end):
        """
        Thuật toán BFS - tìm đường đi ít bước nhất
        Trả về: đường đi và tổng trọng số
        """
        if start not in self.vertices or end not in self.vertices:
            return None, 0
        
        queue = deque([(start, [start], 0)])  # (node, path, total_weight)
        visited = set()
        
        while queue:
            current, path, weight = queue.popleft()
            
            if current in visited:
                continue
            
            visited.add(current)
            
            if current == end:
                return path, weight
            
            # Thêm các đỉnh kề vào queue
            for neighbor, edge_weight in self.graph[current]:
                if neighbor not in visited:
                    new_path = path + [neighbor]
                    new_weight = weight + edge_weight
                    queue.append((neighbor, new_path, new_weight))
        
        return None, 0
    
    def dfs(self, start, end):
        """
        Thuật toán DFS - tìm một đường đi (không đảm bảo tối ưu)
        Trả về: đường đi và tổng trọng số
        """
        if start not in self.vertices or end not in self.vertices:
            return None, 0
        
        visited = set()
        path = []
        
        def dfs_recursive(node, target, current_path, current_weight):
            visited.add(node)
            current_path.append(node)
            
            if node == target:
                return current_path.copy(), current_weight
            
            # Duyệt các đỉnh kề
            for neighbor, edge_weight in self.graph[node]:
                if neighbor not in visited:
                    result_path, result_weight = dfs_recursive(
                        neighbor, target, current_path, current_weight + edge_weight
                    )
                    if result_path:
                        return result_path, result_weight
            
            current_path.pop()
            return None, 0
        
        return dfs_recursive(start, end, [], 0)
    
    def dijkstra(self, start, end):
        """
        Thuật toán Dijkstra - tìm đường đi ngắn nhất theo trọng số
        Sử dụng priority queue (heap)
        """
        if start not in self.vertices or end not in self.vertices:
            return None, 0
        
        # Khởi tạo khoảng cách và đường đi
        distances = {vertex: float('infinity') for vertex in self.vertices}
        previous = {vertex: None for vertex in self.vertices}
        distances[start] = 0
        
        # Priority queue: (distance, vertex)
        pq = [(0, start)]
        visited = set()
        
        while pq:
            current_dist, current = heapq.heappop(pq)
            
            if current in visited:
                continue
            
            visited.add(current)
            
            # Nếu đã đến đích
            if current == end:
                break
            
            # Cập nhật khoảng cách đến các đỉnh kề
            for neighbor, weight in self.graph[current]:
                if neighbor not in visited:
                    new_dist = current_dist + weight
                    
                    if new_dist < distances[neighbor]:
                        distances[neighbor] = new_dist
                        previous[neighbor] = current
                        heapq.heappush(pq, (new_dist, neighbor))
        
        # Xây dựng đường đi
        if distances[end] == float('infinity'):
            return None, 0
        
        path = []
        current = end
        while current is not None:
            path.append(current)
            current = previous[current]
        path.reverse()
        
        return path, distances[end]
    
    def print_graph(self):
        """
        In cấu trúc đồ thị
        """
        print("=== CẤU TRÚC ĐỒ THỊ ===")
        for vertex in sorted(self.vertices):
            neighbors = [(v, w) for v, w in self.graph[vertex]]
            print(f"Đỉnh {vertex}: {neighbors}")
        print()

def main():
    # Tạo đồ thị có trọng số với 10 đỉnh và 15 cạnh
    g = WeightedGraph()
    
    # Thêm các cạnh (đỉnh1, đỉnh2, trọng_số)
    edges = [
        ('A', 'B', 4), ('A', 'C', 2), ('A', 'D', 7),
        ('B', 'C', 1), ('B', 'E', 5), ('B', 'F', 3),
        ('C', 'D', 6), ('C', 'F', 8), ('C', 'G', 4),
        ('D', 'G', 2), ('D', 'H', 9), ('E', 'F', 1),
        ('F', 'G', 3), ('F', 'I', 7), ('G', 'H', 5),
        ('G', 'I', 2), ('H', 'I', 3), ('H', 'J', 4),
        ('I', 'J', 6)
    ]
    
    print("THIẾT KẾ ĐỒ THỊ CÓ TRỌNG SỐ")
    print(f"Số đỉnh: 10 (A-J), Số cạnh: {len(edges)}")
    print()
    
    for u, v, w in edges:
        g.add_edge(u, v, w)
    
    g.print_graph()
    
    # Chọn điểm bắt đầu và kết thúc để test
    start_node = 'A'
    end_node = 'J'
    
    print(f"=== SO SÁNH THUẬT TOÁN: {start_node} → {end_node} ===")
    print()
    
    # 1. Chạy BFS
    print("1. BREADTH-FIRST SEARCH (BFS)")
    print("   Mục tiêu: Tìm đường đi với ÍT BƯỚC NHẤT")
    bfs_path, bfs_weight = g.bfs(start_node, end_node)
    if bfs_path:
        print(f"   Đường đi: {' → '.join(bfs_path)}")
        print(f"   Số bước: {len(bfs_path) - 1}")
        print(f"   Tổng trọng số: {bfs_weight}")
    else:
        print("   Không tìm thấy đường đi")
    print()
    
    # 2. Chạy DFS
    print("2. DEPTH-FIRST SEARCH (DFS)")
    print("   Mục tiêu: Tìm MỘT ĐƯỜNG ĐI bất kỳ")
    dfs_path, dfs_weight = g.dfs(start_node, end_node)
    if dfs_path:
        print(f"   Đường đi: {' → '.join(dfs_path)}")
        print(f"   Số bước: {len(dfs_path) - 1}")
        print(f"   Tổng trọng số: {dfs_weight}")
    else:
        print("   Không tìm thấy đường đi")
    print()
    
    # 3. Chạy Dijkstra
    print("3. DIJKSTRA'S ALGORITHM")
    print("   Mục tiêu: Tìm đường đi với TRỌNG SỐ NHỎ NHẤT")
    dijkstra_path, dijkstra_weight = g.dijkstra(start_node, end_node)
    if dijkstra_path:
        print(f"   Đường đi: {' → '.join(dijkstra_path)}")
        print(f"   Số bước: {len(dijkstra_path) - 1}")
        print(f"   Tổng trọng số: {dijkstra_weight}")
    else:
        print("   Không tìm thấy đường đi")
    print()
    
    # 4. Phân tích và so sánh
    print("=== PHÂN TÍCH VÀ SO SÁNH ===")
    print()
    
    if bfs_path and dijkstra_path:
        print("📊 SO SÁNH BFS vs DIJKSTRA:")
        print(f"   BFS    - Bước: {len(bfs_path)-1:2d}, Trọng số: {bfs_weight:2d}")
        print(f"   Dijkstra - Bước: {len(dijkstra_path)-1:2d}, Trọng số: {dijkstra_weight:2d}")
        print()
        
        if len(bfs_path) < len(dijkstra_path):
            print("✅ BFS tìm được đường đi ÍT BƯỚC HỢN")
        elif len(bfs_path) > len(dijkstra_path):
            print("✅ Dijkstra tìm được đường đi ít bước hơn")
        else:
            print("⚖️  Cả hai có cùng số bước")
        
        if bfs_weight > dijkstra_weight:
            print("✅ Dijkstra tìm được đường đi NHẸ HƠN")
        elif bfs_weight < dijkstra_weight:
            print("✅ BFS tình cờ tìm được đường nhẹ hơn")
        else:
            print("⚖️  Cả hai có cùng trọng số")
    
    print()
    print("=== KẾT LUẬN ===")
    print("🔹 BFS: Đảm bảo ÍT BƯỚC NHẤT (không quan tâm trọng số)")
    print("🔹 DFS: Tìm một đường đi bất kỳ (không tối ưu)")
    print("🔹 Dijkstra: Đảm bảo TỔNG TRỌNG SỐ NHỎ NHẤT")
    print()
    print("💡 Trong thực tế:")
    print("   - Dùng BFS khi muốn ít chuyển đổi/bước nhất")
    print("   - Dùng Dijkstra khi chi phí/trọng số quan trọng")

if __name__ == "__main__":
    main()
````

## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).

- Bi ton lp lch.
- Thit k k thut, machine learning, v.v.

2. Tm gi tr ti thiu ca hm s $g(x, y) = x^2 + y^2$ trong khong $x, y \in [-5, 5]$.

Ngoi ra, chng ta s trc quan ha s thay i ca gi tr fitness qua cc th h.


## V d 1: Ti u ha hm mt bin

### Bi ton

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
# code vo  y nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Bi tp 1: Ti u hm h(x) = sin(x) + cos(x) trong [0, 2] bng GA nh phn 16-bit

# Tham s GA
L = 16                    #  di chui nh phn
pop_size = 50             # Kch thc qun th
generations = 100         # S th h
min_val = 0.0             # Gii hn di ca x
max_val = 2 * np.pi       # Gii hn trn ca x
crossover_rate = 0.8      # Xc sut lai ghp
mutation_rate = 1.0 / L   # Xc sut t bin mi bit

# Khi to qun th (m ha nh phn)
def initialize_population_binary(pop_size, L):
    return [''.join(random.choice('01') for _ in range(L)) for _ in range(pop_size)]

# Gii m chui nh phn sang s thc trong [min_val, max_val]
def decode_genome(genome, min_val, max_val):
    integer = int(genome, 2)
    return min_val + integer * (max_val - min_val) / (2**L - 1)

# Hm mc tiu h(x)
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Chn lc theo Roulette Wheel Selection (bnh xe)
def roulette_wheel_selection(population, fitness_values):
    # m bo  thch nghi khng m
    min_f = min(fitness_values)
    if min_f < 0:
        adjusted = [f - min_f + 1e-6 for f in fitness_values]
    else:
        adjusted = fitness_values[:]
    total = sum(adjusted)
    probs = [f / total for f in adjusted]
    return random.choices(population, weights=probs, k=1)[0]

# Lai ghp mt im
def crossover_one_point(parent1, parent2, rate):
    if random.random() < rate:
        pt = random.randint(1, L-1)
        return parent1[:pt] + parent2[pt:]
    return parent1

# t bin lt bit
def mutate_bitflip(genome, rate):
    new = list(genome)
    for i in range(len(new)):
        if random.random() < rate:
            new[i] = '1' if new[i] == '0' else '0'
    return ''.join(new)

# Thut ton di truyn chnh

def genetic_algorithm_binary(pop_size, generations, L, min_val, max_val, crossover_rate, mutation_rate):
    population = initialize_population_binary(pop_size, L)
    best_history = []
    for gen in range(generations):
        # Gii m v nh gi
        decoded = [decode_genome(g, min_val, max_val) for g in population]
        fitness_vals = [fitness_function(x) for x in decoded]
        best_idx = np.argmax(fitness_vals)
        best_fitness = fitness_vals[best_idx]
        best_x = decoded[best_idx]
        best_history.append(best_fitness)
        print(f'Generation {gen+1}: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
        # To th h mi
        new_pop = []
        while len(new_pop) < pop_size:
            p1 = roulette_wheel_selection(population, fitness_vals)
            p2 = roulette_wheel_selection(population, fitness_vals)
            child = crossover_one_point(p1, p2, crossover_rate)
            child = mutate_bitflip(child, mutation_rate)
            new_pop.append(child)
        population = new_pop
    # Kt qu cui cng
    decoded = [decode_genome(g, min_val, max_val) for g in population]
    fitness_vals = [fitness_function(x) for x in decoded]
    best_idx = np.argmax(fitness_vals)
    best_x = decoded[best_idx]
    best_fitness = fitness_vals[best_idx]
    # V biu 
    plt.plot(range(1, generations+1), best_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) ti u')
    plt.title('GA Binary: Qu trnh ti u ha h(x)')
    plt.show()
    print(f'\nKt qu cui cng: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
    return best_x, best_fitness

# Chy GA
best_x, best_h = genetic_algorithm_binary(pop_size, generations, L, min_val, max_val, crossover_rate, mutation_rate)
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
# code vo y nhe
import numpy as np
import random
import matplotlib.pyplot as plt

"""
Lab3 - Bi tp 2: Ti u hm h(x) = sin(x) + cos(x) bng GA nh phn 16bit
Tham kho khung GA  v d 1 trong lab3.ipynb, thay i:
   mutation_rate = 0.05 (5%)
   Gi nguyn Roulette Wheel, onepoint crossover 0.8
   nh gi so snh vi nghim chnh xc 2 ti x = /4
"""

# ====================== Tham s GA ======================
L = 16                    # s bit ca nhim sc th
POP_SIZE = 50             # kch thc qun th
GENERATIONS = 100         # s th h
MIN_VAL = 0.0             # gii hn di ca x
MAX_VAL = 2 * np.pi       # gii hn trn ca x
CROSSOVER_RATE = 0.8      # xc sut lai ghp
MUTATION_RATE = 0.05      # xc sut t bin mi bit   (Bi tp 2 yu cu 5%)

# ====================== Hm tin ch ======================

def initialize_population(pop_size: int, length: int):
    """Khi to qun th nh phn ngu nhin"""
    return ["".join(random.choice("01") for _ in range(length)) for _ in range(pop_size)]


def decode(genome: str, min_val: float, max_val: float):
    """Gii m chui nh phn sang s thc trong [min_val, max_val]"""
    integer = int(genome, 2)
    return min_val + integer * (max_val - min_val) / (2 ** L - 1)


def fitness_function(x: float):
    """Fitness = h(x) = sin(x) + cos(x)"""
    return np.sin(x) + np.cos(x)


def roulette_select(population, fitness_vals):
    """Chn 1 c th bng Roulette Wheel Selection"""
    min_f = min(fitness_vals)
    adjusted = [f - min_f + 1e-6 for f in fitness_vals]  # m bo dng
    total = sum(adjusted)
    probs = [f / total for f in adjusted]
    return random.choices(population, weights=probs, k=1)[0]


def one_point_crossover(p1: str, p2: str, rate: float):
    if random.random() < rate:
        pt = random.randint(1, L - 1)
        return p1[:pt] + p2[pt:]
    return p1


def bitflip_mutation(genome: str, rate: float):
    genome_list = list(genome)
    for i in range(len(genome_list)):
        if random.random() < rate:
            genome_list[i] = "1" if genome_list[i] == "0" else "0"
    return "".join(genome_list)

# ====================== GA chnh ======================

def genetic_algorithm():
    population = initialize_population(POP_SIZE, L)
    best_history = []

    for gen in range(GENERATIONS):
        decoded = [decode(g, MIN_VAL, MAX_VAL) for g in population]
        fitness_vals = [fitness_function(x) for x in decoded]

        # Ghi nhn c th tt nht
        best_idx = int(np.argmax(fitness_vals))
        best_x = decoded[best_idx]
        best_fit = fitness_vals[best_idx]
        best_history.append(best_fit)

        print(f"Generation {gen+1:3d}: x = {best_x:.6f}, h(x) = {best_fit:.6f}")

        # To qun th mi
        new_population = []
        while len(new_population) < POP_SIZE:
            p1 = roulette_select(population, fitness_vals)
            p2 = roulette_select(population, fitness_vals)
            child = one_point_crossover(p1, p2, CROSSOVER_RATE)
            child = bitflip_mutation(child, MUTATION_RATE)
            new_population.append(child)
        population = new_population

    # Kt qu cui
    decoded = [decode(g, MIN_VAL, MAX_VAL) for g in population]
    fitness_vals = [fitness_function(x) for x in decoded]
    best_idx = int(np.argmax(fitness_vals))
    best_x = decoded[best_idx]
    best_fit = fitness_vals[best_idx]

    # V biu 
    plt.figure()
    plt.plot(range(1, GENERATIONS + 1), best_history)
    plt.xlabel("Th h")
    plt.ylabel("Gi tr h(x) ti u")
    plt.title("GA Binary 16bit  Bi tp 2")
    plt.show()

    # So snh vi nghim chnh xc
    exact_max = np.sqrt(2)
    exact_x = np.pi / 4
    err = abs(best_fit - exact_max)
    print("\n================ KT QU CUI ================")
    print(f"GA: x* = {best_x:.6f}, h(x*) = {best_fit:.6f}")
    print(f"Exact: x_opt = {exact_x:.6f}, h(x_opt) = {exact_max:.6f}")
    print(f"|GA  Exact| = {err:.6e}\n")

    return best_x, best_fit


if __name__ == "__main__":
    genetic_algorithm()
````

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
# code vo y 
import numpy as np
import random
import matplotlib.pyplot as plt

# Lab3 - Bi tp 3: Phn tch tc  hi t khi thay i tham s GA
# - Chy nhiu ln  n nh kt qu
# - Tnh generation trung bnh  t ngng

# ====================== Cu hnh chung ======================
L = 16
MIN_VAL = 0.0
MAX_VAL = 2 * np.pi
POP_SIZE_DEFAULT = 50
CROSSOVER_DEFAULT = 0.8
MUTATION_DEFAULT = 0.05
RUNS = 10                    # s ln lp cho mi tham s
THRESHOLD_RATIO = 0.95       # ngng hi t: 95% gi tr ti u 2
exact_max = np.sqrt(2)

# ====================== Hm GA ======================
def initialize_population(pop_size):
    return [''.join(random.choice('01') for _ in range(L)) for _ in range(pop_size)]

def decode(genome):
    integer = int(genome, 2)
    return MIN_VAL + integer * (MAX_VAL - MIN_VAL) / (2**L - 1)

def fitness(x):
    return np.sin(x) + np.cos(x)

def roulette_select(pop, fitness_vals):
    min_f = min(fitness_vals)
    adj = [f - min_f + 1e-6 for f in fitness_vals]
    total = sum(adj)
    probs = [f/total for f in adj]
    return random.choices(pop, weights=probs, k=1)[0]

def one_point_crossover(p1, p2, rate):
    if random.random() < rate:
        pt = random.randint(1, L-1)
        return p1[:pt] + p2[pt:]
    return p1

def bitflip_mutation(genome, rate):
    g = list(genome)
    for i in range(len(g)):
        if random.random() < rate:
            g[i] = '1' if g[i]=='0' else '0'
    return ''.join(g)

def run_ga(pop_size, crossover_rate, mutation_rate, generations):
    pop = initialize_population(pop_size)
    best_hist = []
    for gen in range(generations):
        decoded = [decode(g) for g in pop]
        fvals = [fitness(x) for x in decoded]
        best_hist.append(max(fvals))
        new_pop = []
        while len(new_pop) < pop_size:
            p1 = roulette_select(pop, fvals)
            p2 = roulette_select(pop, fvals)
            child = one_point_crossover(p1, p2, crossover_rate)
            child = bitflip_mutation(child, mutation_rate)
            new_pop.append(child)
        pop = new_pop
    return best_hist

# ====================== Th nghim nh lng ======================
# Thm phn tch cho 4 tham s: pop_size, generations, mutation_rate, crossover_rate
param_sets = {
    'pop_size': [20, 50, 100],
    'generations': [50, 100, 150],
    'mutation_rate': [0.01, 0.05, 0.1],
    'crossover_rate': [0.6, 0.8, 1.0]
}

quantitative = {}
for param, values in param_sets.items():
    quantitative[param] = {}
    plt.figure(figsize=(8, 5))
    for v in values:
        all_runs = []
        conv_gens = []
        for _ in range(RUNS):
            # Gi run_ga vi tham s ang thay i
            if param == 'pop_size':
                hist = run_ga(v, CROSSOVER_DEFAULT, MUTATION_DEFAULT, 100)
            elif param == 'generations':
                hist = run_ga(POP_SIZE_DEFAULT, CROSSOVER_DEFAULT, MUTATION_DEFAULT, v)
            elif param == 'mutation_rate':
                hist = run_ga(POP_SIZE_DEFAULT, CROSSOVER_DEFAULT, v, 100)
            else:  # crossover_rate
                hist = run_ga(POP_SIZE_DEFAULT, v, MUTATION_DEFAULT, 100)
            all_runs.append(hist)
            # tm generation u tin t threshold
            gens = [i for i, val in enumerate(hist, 1) if val >= THRESHOLD_RATIO * exact_max]
            conv_gens.append(gens[0] if gens else len(hist))
        mean_hist = np.mean(all_runs, axis=0)
        std_hist = np.std(all_runs, axis=0)
        avg_conv = np.mean(conv_gens)
        std_conv = np.std(conv_gens)
        quantitative[param][v] = (avg_conv, std_conv)
        # v  th mean  std
        gens_range = range(1, len(mean_hist)+1)
        plt.plot(gens_range, mean_hist, label=f'{param}={v}, conv={avg_conv:.1f}{std_conv:.1f}')
        plt.fill_between(gens_range,
                         mean_hist-std_hist,
                         mean_hist+std_hist,
                         alpha=0.2)
    plt.title(f'Convergence: nh hng ca {param}')
    plt.xlabel('Th h')
    plt.ylabel('Fitness tt nht')
    plt.legend()
    plt.show()

# ====================== Nhn xt c th ======================
print('=== Nhn xt nh lng tc  hi t ===')
for param, stats in quantitative.items():
    print(f'* {param}:')
    for v, (avg, sd) in stats.items():
        print(f'  - {v}: trung bnh hi t  ~{avg:.1f} gen ({sd:.1f})')

print('\nNhn xt chung:')
print('- Qun th ln (pop_size=100) n nh nhng cn nhiu th h hn.')
print('- Tng generations gip ci thin  chnh xc hi t nhng tng thi gian.')
print('- Mutation_rate 0.01 hi t nhanh nhng d mc kt; 0.1 a dng hn nhng chm hn.')
print('- Crossover_rate cao (1.0) nhanh t ngng nhng dao ng; thp (0.6) n nh nhng chm.')
````

# Ci tin thut ton  Thm phng php la chn khc v so snh hiu qu

## Mc tiu

- Thm phng php la chn mi vo thut ton di truyn, v d nh **Roulette Wheel Selection** (La chn bnh xe con quay).
- So snh hiu qu ca phng php la chn mi vi phng php c (v d: Chn ngu nhin, Tournament Selection, v.v.).
- Phn tch nh hng ca phng php la chn n tc  hi t v cht lng nghim.

---

##  Cc phng php la chn ph bin

| Phng php                | M t                                                                                   | u im                              | Nhc im                          |
|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|
| Chn ngu nhin (Random)   | Chn c th ngu nhin trong qun th                                                   | n gin, nhanh                    | Khng u tin c th tt hn      |
| Roulette Wheel Selection    | Chn theo xc sut t l thun vi  thch nghi (fitness)                              | D dng, u tin c th tt        | D b nh hng bi c th qu mnh|
| Tournament Selection        | Chn ngu nhin mt s c th, chn c th tt nht trong nhm                          | Kim sot a dng tt              | Cn chn kch thc nhm ph hp  |

---

##   tng ci tin

1. Ci t thm phng php **Roulette Wheel Selection** trong thut ton di truyn.
2. Chy th vi cng b tham s trn cc phng php la chn khc nhau.
3. So snh kt qu v:
   - Tc  hi t (s th h cn  t nghim tt).
   - Gi tr nghim tt nht t c.
4. Phn tch u  nhc im tng phng php trong thc t.

---

##  Kt qu v phn tch

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

---


````python
# code vo y nh
import numpy as np
import random
import matplotlib.pyplot as plt

# 1. Hm mc tiu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# 2. Khi to qun th (reuse t lab3: initialize_population nhng vi gii hn [0,2])
def initialize_population(pop_size, min_val=0, max_val=2*np.pi):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# 3. Cc phng php chn lc
def select_random(population, fitness_values):
    # Chn 1 c th ngu nhin
    return random.choice(population)

def select_tournament(population, fitness_values, tournament_size=3):
    # reuse t lab3
    contenders = random.sample(range(len(population)), tournament_size)
    best = max(contenders, key=lambda i: fitness_values[i])
    return population[best]

def select_roulette(population, fitness_values):
    # Chn theo xc sut t l thun vi fitness (di min thnh dng)
    min_f = min(fitness_values)
    shifted = [f - min_f + 1e-6 for f in fitness_values]
    total = sum(shifted)
    probs = [f/total for f in shifted]
    return random.choices(population, weights=probs, k=1)[0]

# 4. Lai ghp & t bin: reuse from lab3 (arithmetic crossover + mutate real)
def crossover(parent1, parent2, rate=0.8):
    if random.random() < rate:
        return (parent1 + parent2) / 2
    return parent1

def mutate(individual, rate=0.1, min_val=0, max_val=2*np.pi):
    if random.random() < rate:
        return random.uniform(min_val, max_val)
    return individual

# 5. Hm GA tng qut vi tham s chn lc
def genetic_algorithm(selection_fn, pop_size=50, generations=100,
                      crossover_rate=0.8, mutation_rate=0.1):
    pop = initialize_population(pop_size)
    best_history = []
    for gen in range(generations):
        fitness_vals = [fitness_function(x) for x in pop]
        best_history.append(max(fitness_vals))
        new_pop = []
        for _ in range(pop_size):
            p1 = selection_fn(pop, fitness_vals)
            p2 = selection_fn(pop, fitness_vals)
            child = crossover(p1, p2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_pop.append(child)
        pop = new_pop
    return best_history

# 6. Chy th & v so snh
methods = {
    "Random": select_random,
    "Tournament": select_tournament,
    "Roulette": select_roulette
}
plt.figure(figsize=(8,5))
for name, fn in methods.items():
    hist = genetic_algorithm(fn, pop_size=100, generations=100,
                             crossover_rate=0.8, mutation_rate=0.05)
    plt.plot(hist, label=name)
plt.xlabel('Th h')
plt.ylabel('Fitness tt nht')
plt.title('So snh tc  hi t ca cc phng php chn lc')
plt.legend()
plt.show()
````

# Bi tp 5
# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

---

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
#code vo y
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  #  v 3D

# --- 1. Hm mc tiu 3D ---
def fitness_function_3d(x, y, z):
    #  y k(x,y,z) = x^2 + y^2 + z^2, ta mun MIN nn i du  GA MAX
    return -(x**2 + y**2 + z**2)

# --- 2. Khi to qun th 3D ---
def initialize_population_3d(pop_size, min_val=-5, max_val=5):
    return [
        (random.uniform(min_val, max_val),
         random.uniform(min_val, max_val),
         random.uniform(min_val, max_val))
        for _ in range(pop_size)
    ]

# --- 3. Chn lc Tournament (reuse) ---
def select_parents_3d(pop, fitness_vals, tournament_size=3):
    contenders = random.sample(range(len(pop)), tournament_size)
    best_idx = max(contenders, key=lambda i: fitness_vals[i])
    return pop[best_idx]

# --- 4. Lai ghp & t bin 3D ---
def crossover_3d(p1, p2, rate=0.8):
    if random.random() < rate:
        return (
            (p1[0] + p2[0]) / 2,
            (p1[1] + p2[1]) / 2,
            (p1[2] + p2[2]) / 2
        )
    return p1

def mutate_3d(ind, rate=0.1, min_val=-5, max_val=5):
    x, y, z = ind
    if random.random() < rate:
        x = random.uniform(min_val, max_val)
    if random.random() < rate:
        y = random.uniform(min_val, max_val)
    if random.random() < rate:
        z = random.uniform(min_val, max_val)
    return (x, y, z)

# --- 5. GA 3D, tr v danh sch qun th qua tng th h ---
def genetic_algorithm_3d(pop_size=100, generations=50,
                         crossover_rate=0.8, mutation_rate=0.1):
    population = initialize_population_3d(pop_size)
    history = [population]  # lu qun th th h 0
    for gen in range(generations):
        fitness_vals = [fitness_function_3d(x,y,z) for x,y,z in population]
        new_pop = []
        for _ in range(pop_size):
            p1 = select_parents_3d(population, fitness_vals)
            p2 = select_parents_3d(population, fitness_vals)
            child = crossover_3d(p1, p2, crossover_rate)
            child = mutate_3d(child, mutation_rate)
            new_pop.append(child)
        population = new_pop
        history.append(population)
    return history

# --- 6. Chy GA v ly cc th h mu  v ---
hist = genetic_algorithm_3d(pop_size=200, generations=60,
                            crossover_rate=0.8, mutation_rate=0.05)

# Chn gen 0, gen gia v gen cui  trc quan
gens_to_plot = [0, len(hist)//2, len(hist)-1]
fig = plt.figure(figsize=(12,4))
for i, g in enumerate(gens_to_plot):
    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    pop = hist[g]
    xs, ys, zs = zip(*pop)
    ax.scatter(xs, ys, zs, s=10, alpha=0.6)
    ax.set_title(f'Generation {g}')
    ax.set_xlim(-5,5); ax.set_ylim(-5,5); ax.set_zlim(-5,5)
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')

plt.suptitle('Scatter plot qun th qua cc th h')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

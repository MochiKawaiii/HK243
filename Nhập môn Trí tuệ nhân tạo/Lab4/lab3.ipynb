## Lý thuyết
Thuật toán di truyền (Genetic Algorithm - GA) là một phương pháp tìm kiếm dựa trên cơ chế tiến hóa tự nhiên, lấy cảm hứng từ thuyết tiến hóa của Darwin. GA được sử dụng để tìm lời giải gần tối ưu cho các bài toán tối ưu hóa phức tạp.

### Các bước chính của thuật toán:
1. **Khởi tạo quần thể**: Tạo ngẫu nhiên một tập hợp các cá thể (giải pháp tiềm năng).
2. **Đánh giá độ thích nghi (Fitness Evaluation)**: Tính giá trị hàm mục tiêu cho mỗi cá thể.
3. **Lựa chọn (Selection)**: Chọn các cá thể tốt nhất để lai ghép (thường sử dụng các phương pháp như Tournament Selection hoặc Roulette Wheel).
4. **Lai ghép (Crossover)**: Kết hợp hai cá thể để tạo ra cá thể mới.
5. **Đột biến (Mutation)**: Thay đổi ngẫu nhiên một số gen để duy trì tính đa dạng.
6. **Lặp lại**: Quay lại bước 2 cho đến khi đạt điều kiện dừng (số thế hệ hoặc giá trị fitness mong muốn).

## Ví dụ 1: Tối ưu hóa hàm một biến

### Bài toán  
Tìm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
đạt giá trị lớn nhất trong khoảng  
$$
x \in [0, 10].
$$

# Mã giả
```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection:
        - Chọn ngẫu nhiên tournament_size cá thể trong quần thể
        - Chọn cá thể có fitness cao nhất làm cha mẹ 1
    Chọn cha mẹ 2 bằng Tournament Selection tương tự
    
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới có giá trị trung bình của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
            
    Thực hiện đột biến cho con mới với xác suất mutation_rate:
        Nếu đột biến:
            Gán con mới giá trị ngẫu nhiên trong [min_val, max_val]
            
    Thêm con mới vào quần thể mới
    
Cập nhật quần thể hiện tại = quần thể mới
In ra cá thể tốt nhất và giá trị fitness tương ứng của thế hệ


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

````python
# Hàm mục tiêu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # Sử dụng dấu âm vì GA tìm max

# Khởi tạo quần thể
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# Lựa chọn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# Đột biến
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thuật toán di truyền
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyển về giá trị thực
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Thế hệ {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị g(x, y) tốt nhất')
    plt.title('Sự thay đổi của g(x, y) qua các thế hệ')
    plt.show()
    
    return best_individual, -best_fitness

# Chạy thuật toán
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKết quả cuối cùng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bài tập 1
# Thay đổi hàm mục tiêu**: Thử áp dụng thuật toán cho hàm $h(x) = sin(x) + cos(x)$ trong khoảng $x \in [0, 2\pi]$. So sánh kết quả với nghiệm chính xác.

## Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

## Mục tiêu
- Tối ưu hàm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong miền:
  $$
  x \in [0, 2\pi]
  $$
- Tìm giá trị **x** sao cho **h(x)** đạt giá trị lớn nhất.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

## Ý tưởng thuật toán di truyền (GA)

1. **Mã hóa**: Biểu diễn giá trị \( x \) bằng chuỗi nhị phân dài 16 bit.  
2. **Giải mã**: Chuyển từ chuỗi nhị phân sang giá trị thực trong khoảng \([0, 2\pi]\).  
3. **Fitness**: Sử dụng chính giá trị \( h(x) \) làm độ thích nghi.  
4. **Chọn lọc**: Roulette Wheel Selection.  
5. **Lai ghép**: Lai 1 điểm với xác suất cao.  
6. **Đột biến**: Lật bit với xác suất thấp.

---


````python
# code vào  đây nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Bài tập 1: Tối ưu hàm h(x) = sin(x) + cos(x) trong [0, 2π] bằng GA nhị phân 16-bit

# Tham số GA
L = 16                    # Độ dài chuỗi nhị phân
pop_size = 50             # Kích thước quần thể
generations = 100         # Số thế hệ
min_val = 0.0             # Giới hạn dưới của x
max_val = 2 * np.pi       # Giới hạn trên của x
crossover_rate = 0.8      # Xác suất lai ghép
mutation_rate = 1.0 / L   # Xác suất đột biến mỗi bit

# Khởi tạo quần thể (mã hóa nhị phân)
def initialize_population_binary(pop_size, L):
    return [''.join(random.choice('01') for _ in range(L)) for _ in range(pop_size)]

# Giải mã chuỗi nhị phân sang số thực trong [min_val, max_val]
def decode_genome(genome, min_val, max_val):
    integer = int(genome, 2)
    return min_val + integer * (max_val - min_val) / (2**L - 1)

# Hàm mục tiêu h(x)
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Chọn lọc theo Roulette Wheel Selection (bánh xe)
def roulette_wheel_selection(population, fitness_values):
    # Đảm bảo độ thích nghi không âm
    min_f = min(fitness_values)
    if min_f < 0:
        adjusted = [f - min_f + 1e-6 for f in fitness_values]
    else:
        adjusted = fitness_values[:]
    total = sum(adjusted)
    probs = [f / total for f in adjusted]
    return random.choices(population, weights=probs, k=1)[0]

# Lai ghép một điểm
def crossover_one_point(parent1, parent2, rate):
    if random.random() < rate:
        pt = random.randint(1, L-1)
        return parent1[:pt] + parent2[pt:]
    return parent1

# Đột biến lật bit
def mutate_bitflip(genome, rate):
    new = list(genome)
    for i in range(len(new)):
        if random.random() < rate:
            new[i] = '1' if new[i] == '0' else '0'
    return ''.join(new)

# Thuật toán di truyền chính

def genetic_algorithm_binary(pop_size, generations, L, min_val, max_val, crossover_rate, mutation_rate):
    population = initialize_population_binary(pop_size, L)
    best_history = []
    for gen in range(generations):
        # Giải mã và đánh giá
        decoded = [decode_genome(g, min_val, max_val) for g in population]
        fitness_vals = [fitness_function(x) for x in decoded]
        best_idx = np.argmax(fitness_vals)
        best_fitness = fitness_vals[best_idx]
        best_x = decoded[best_idx]
        best_history.append(best_fitness)
        print(f'Generation {gen+1}: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
        # Tạo thế hệ mới
        new_pop = []
        while len(new_pop) < pop_size:
            p1 = roulette_wheel_selection(population, fitness_vals)
            p2 = roulette_wheel_selection(population, fitness_vals)
            child = crossover_one_point(p1, p2, crossover_rate)
            child = mutate_bitflip(child, mutation_rate)
            new_pop.append(child)
        population = new_pop
    # Kết quả cuối cùng
    decoded = [decode_genome(g, min_val, max_val) for g in population]
    fitness_vals = [fitness_function(x) for x in decoded]
    best_idx = np.argmax(fitness_vals)
    best_x = decoded[best_idx]
    best_fitness = fitness_vals[best_idx]
    # Vẽ biểu đồ
    plt.plot(range(1, generations+1), best_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tối ưu')
    plt.title('GA Binary: Quá trình tối ưu hóa h(x)')
    plt.show()
    print(f'\nKết quả cuối cùng: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
    return best_x, best_fitness

# Chạy GA
best_x, best_h = genetic_algorithm_binary(pop_size, generations, L, min_val, max_val, crossover_rate, mutation_rate)
````

---

##  Ý tưởng thuật toán di truyền (GA)

| Thành phần        | Chi tiết                                                                 |
|------------------|-------------------------------------------------------------------------|
| **Mã hóa**        | Dùng chuỗi nhị phân 16 bit để biểu diễn giá trị $x \in [0, 2\pi]$         |
| **Giải mã**       | Ánh xạ nhị phân sang giá trị thực: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chính là $h(x) = \sin(x) + \cos(x)$                                     |
| **Chọn lọc**      | Roulette Wheel Selection                                                |
| **Lai ghép**      | 1 điểm cắt với xác suất 0.8                                              |
| **Đột biến**      | Xác suất 5% (0.05)                                                       |

````python
# code vào đây nhe
import numpy as np
import random
import matplotlib.pyplot as plt

"""
Lab3 - Bài tập 2: Tối ưu hàm h(x) = sin(x) + cos(x) bằng GA nhị phân 16‑bit
Tham khảo khung GA ở ví dụ 1 trong lab3.ipynb, thay đổi:
  • mutation_rate = 0.05 (5%)
  • Giữ nguyên Roulette Wheel, one‑point crossover 0.8
  • Đánh giá so sánh với nghiệm chính xác √2 tại x = π/4
"""

# ====================== Tham số GA ======================
L = 16                    # số bit của nhiễm sắc thể
POP_SIZE = 50             # kích thước quần thể
GENERATIONS = 100         # số thế hệ
MIN_VAL = 0.0             # giới hạn dưới của x
MAX_VAL = 2 * np.pi       # giới hạn trên của x
CROSSOVER_RATE = 0.8      # xác suất lai ghép
MUTATION_RATE = 0.05      # xác suất đột biến mỗi bit   (Bài tập 2 yêu cầu 5%)

# ====================== Hàm tiện ích ======================

def initialize_population(pop_size: int, length: int):
    """Khởi tạo quần thể nhị phân ngẫu nhiên"""
    return ["".join(random.choice("01") for _ in range(length)) for _ in range(pop_size)]


def decode(genome: str, min_val: float, max_val: float):
    """Giải mã chuỗi nhị phân sang số thực trong [min_val, max_val]"""
    integer = int(genome, 2)
    return min_val + integer * (max_val - min_val) / (2 ** L - 1)


def fitness_function(x: float):
    """Fitness = h(x) = sin(x) + cos(x)"""
    return np.sin(x) + np.cos(x)


def roulette_select(population, fitness_vals):
    """Chọn 1 cá thể bằng Roulette Wheel Selection"""
    min_f = min(fitness_vals)
    adjusted = [f - min_f + 1e-6 for f in fitness_vals]  # đảm bảo dương
    total = sum(adjusted)
    probs = [f / total for f in adjusted]
    return random.choices(population, weights=probs, k=1)[0]


def one_point_crossover(p1: str, p2: str, rate: float):
    if random.random() < rate:
        pt = random.randint(1, L - 1)
        return p1[:pt] + p2[pt:]
    return p1


def bitflip_mutation(genome: str, rate: float):
    genome_list = list(genome)
    for i in range(len(genome_list)):
        if random.random() < rate:
            genome_list[i] = "1" if genome_list[i] == "0" else "0"
    return "".join(genome_list)

# ====================== GA chính ======================

def genetic_algorithm():
    population = initialize_population(POP_SIZE, L)
    best_history = []

    for gen in range(GENERATIONS):
        decoded = [decode(g, MIN_VAL, MAX_VAL) for g in population]
        fitness_vals = [fitness_function(x) for x in decoded]

        # Ghi nhận cá thể tốt nhất
        best_idx = int(np.argmax(fitness_vals))
        best_x = decoded[best_idx]
        best_fit = fitness_vals[best_idx]
        best_history.append(best_fit)

        print(f"Generation {gen+1:3d}: x = {best_x:.6f}, h(x) = {best_fit:.6f}")

        # Tạo quần thể mới
        new_population = []
        while len(new_population) < POP_SIZE:
            p1 = roulette_select(population, fitness_vals)
            p2 = roulette_select(population, fitness_vals)
            child = one_point_crossover(p1, p2, CROSSOVER_RATE)
            child = bitflip_mutation(child, MUTATION_RATE)
            new_population.append(child)
        population = new_population

    # Kết quả cuối
    decoded = [decode(g, MIN_VAL, MAX_VAL) for g in population]
    fitness_vals = [fitness_function(x) for x in decoded]
    best_idx = int(np.argmax(fitness_vals))
    best_x = decoded[best_idx]
    best_fit = fitness_vals[best_idx]

    # Vẽ biểu đồ
    plt.figure()
    plt.plot(range(1, GENERATIONS + 1), best_history)
    plt.xlabel("Thế hệ")
    plt.ylabel("Giá trị h(x) tối ưu")
    plt.title("GA Binary 16‑bit – Bài tập 2")
    plt.show()

    # So sánh với nghiệm chính xác
    exact_max = np.sqrt(2)
    exact_x = np.pi / 4
    err = abs(best_fit - exact_max)
    print("\n================ KẾT QUẢ CUỐI ================")
    print(f"GA: x* = {best_x:.6f}, h(x*) = {best_fit:.6f}")
    print(f"Exact: x_opt = {exact_x:.6f}, h(x_opt) = {exact_max:.6f}")
    print(f"|GA − Exact| = {err:.6e}\n")

    return best_x, best_fit


if __name__ == "__main__":
    genetic_algorithm()
````

# Thay đổi tham số thuật toán di truyền và phân tích tốc độ hội tụ

**Mục tiêu:** Thử thay đổi các tham số chính của thuật toán di truyền gồm:

- Kích thước quần thể: `pop_size`
- Số thế hệ: `generations`
- Tỷ lệ đột biến: `mutation_rate`
- Tỷ lệ lai ghép: `crossover_rate`

và phân tích ảnh hưởng của từng tham số đến tốc độ hội tụ của thuật toán.

---

## Các tham số và vai trò

| Tham số         | Ý nghĩa                                  | Ảnh hưởng chính                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | Số cá thể trong quần thể                 | Quần thể lớn giúp khám phá không gian tìm kiếm tốt hơn nhưng tốn nhiều thời gian tính toán. Quần thể nhỏ hội tụ nhanh nhưng dễ bị kẹt cực trị cục bộ. |
| `generations`   | Số vòng lặp của thuật toán               | Số thế hệ càng nhiều thì thuật toán có nhiều cơ hội cải thiện nghiệm, tuy nhiên thời gian chạy cũng tăng. |
| `mutation_rate` | Xác suất đột biến mỗi bit                | Đột biến cao giúp tăng đa dạng gene, tránh bị kẹt trong cực trị cục bộ, nhưng nếu quá cao làm quá trình tìm kiếm nhiễu loạn. |
| `crossover_rate`| Xác suất lai ghép giữa hai cá thể       | Lai ghép giúp phối hợp gen tốt, tăng khả năng tìm nghiệm tốt. Nếu quá thấp, quần thể kém đa dạng; quá cao có thể phá hủy cấu trúc tốt của cá thể. |

---

## Phân tích ảnh hưởng

- **`pop_size`**: Kích thước quần thể lớn giúp tìm kiếm toàn cục tốt hơn nhưng chậm hơn, kích thước nhỏ làm thuật toán hội tụ nhanh nhưng dễ mắc kẹt.
- **`generations`**: Số thế hệ lớn tăng khả năng cải tiến kết quả nhưng làm tăng thời gian tính toán.
- **`mutation_rate`**: Tỷ lệ đột biến vừa phải giúp duy trì đa dạng di truyền, tránh kẹt cục bộ; quá cao hoặc quá thấp đều làm giảm hiệu quả.
- **`crossover_rate`**: Tỷ lệ lai ghép hợp lý giúp khai thác tốt thông tin di truyền, tăng hiệu quả hội tụ.

---




````python
# code vào đây 
import numpy as np
import random
import matplotlib.pyplot as plt

# Lab3 - Bài tập 3: Phân tích tốc độ hội tụ khi thay đổi tham số GA
# - Chạy nhiều lần để ổn định kết quả
# - Tính generation trung bình để đạt ngưỡng

# ====================== Cấu hình chung ======================
L = 16
MIN_VAL = 0.0
MAX_VAL = 2 * np.pi
POP_SIZE_DEFAULT = 50
CROSSOVER_DEFAULT = 0.8
MUTATION_DEFAULT = 0.05
RUNS = 10                    # số lần lặp cho mỗi tham số
THRESHOLD_RATIO = 0.95       # ngưỡng hội tụ: 95% giá trị tối ưu √2
exact_max = np.sqrt(2)

# ====================== Hàm GA ======================
def initialize_population(pop_size):
    return [''.join(random.choice('01') for _ in range(L)) for _ in range(pop_size)]

def decode(genome):
    integer = int(genome, 2)
    return MIN_VAL + integer * (MAX_VAL - MIN_VAL) / (2**L - 1)

def fitness(x):
    return np.sin(x) + np.cos(x)

def roulette_select(pop, fitness_vals):
    min_f = min(fitness_vals)
    adj = [f - min_f + 1e-6 for f in fitness_vals]
    total = sum(adj)
    probs = [f/total for f in adj]
    return random.choices(pop, weights=probs, k=1)[0]

def one_point_crossover(p1, p2, rate):
    if random.random() < rate:
        pt = random.randint(1, L-1)
        return p1[:pt] + p2[pt:]
    return p1

def bitflip_mutation(genome, rate):
    g = list(genome)
    for i in range(len(g)):
        if random.random() < rate:
            g[i] = '1' if g[i]=='0' else '0'
    return ''.join(g)

def run_ga(pop_size, crossover_rate, mutation_rate, generations):
    pop = initialize_population(pop_size)
    best_hist = []
    for gen in range(generations):
        decoded = [decode(g) for g in pop]
        fvals = [fitness(x) for x in decoded]
        best_hist.append(max(fvals))
        new_pop = []
        while len(new_pop) < pop_size:
            p1 = roulette_select(pop, fvals)
            p2 = roulette_select(pop, fvals)
            child = one_point_crossover(p1, p2, crossover_rate)
            child = bitflip_mutation(child, mutation_rate)
            new_pop.append(child)
        pop = new_pop
    return best_hist

# ====================== Thử nghiệm định lượng ======================
# Thêm phân tích cho 4 tham số: pop_size, generations, mutation_rate, crossover_rate
param_sets = {
    'pop_size': [20, 50, 100],
    'generations': [50, 100, 150],
    'mutation_rate': [0.01, 0.05, 0.1],
    'crossover_rate': [0.6, 0.8, 1.0]
}

quantitative = {}
for param, values in param_sets.items():
    quantitative[param] = {}
    plt.figure(figsize=(8, 5))
    for v in values:
        all_runs = []
        conv_gens = []
        for _ in range(RUNS):
            # Gọi run_ga với tham số đang thay đổi
            if param == 'pop_size':
                hist = run_ga(v, CROSSOVER_DEFAULT, MUTATION_DEFAULT, 100)
            elif param == 'generations':
                hist = run_ga(POP_SIZE_DEFAULT, CROSSOVER_DEFAULT, MUTATION_DEFAULT, v)
            elif param == 'mutation_rate':
                hist = run_ga(POP_SIZE_DEFAULT, CROSSOVER_DEFAULT, v, 100)
            else:  # crossover_rate
                hist = run_ga(POP_SIZE_DEFAULT, v, MUTATION_DEFAULT, 100)
            all_runs.append(hist)
            # tìm generation đầu tiên đạt threshold
            gens = [i for i, val in enumerate(hist, 1) if val >= THRESHOLD_RATIO * exact_max]
            conv_gens.append(gens[0] if gens else len(hist))
        mean_hist = np.mean(all_runs, axis=0)
        std_hist = np.std(all_runs, axis=0)
        avg_conv = np.mean(conv_gens)
        std_conv = np.std(conv_gens)
        quantitative[param][v] = (avg_conv, std_conv)
        # vẽ đồ thị mean ± std
        gens_range = range(1, len(mean_hist)+1)
        plt.plot(gens_range, mean_hist, label=f'{param}={v}, conv={avg_conv:.1f}±{std_conv:.1f}')
        plt.fill_between(gens_range,
                         mean_hist-std_hist,
                         mean_hist+std_hist,
                         alpha=0.2)
    plt.title(f'Convergence: ảnh hưởng của {param}')
    plt.xlabel('Thế hệ')
    plt.ylabel('Fitness tốt nhất')
    plt.legend()
    plt.show()

# ====================== Nhận xét cụ thể ======================
print('=== Nhận xét định lượng tốc độ hội tụ ===')
for param, stats in quantitative.items():
    print(f'* {param}:')
    for v, (avg, sd) in stats.items():
        print(f'  - {v}: trung bình hội tụ ở ~{avg:.1f} gen (±{sd:.1f})')

print('\nNhận xét chung:')
print('- Quần thể lớn (pop_size=100) ổn định nhưng cần nhiều thế hệ hơn.')
print('- Tăng generations giúp cải thiện độ chính xác hội tụ nhưng tăng thời gian.')
print('- Mutation_rate 0.01 hội tụ nhanh nhưng dễ mắc kẹt; 0.1 đa dạng hơn nhưng chậm hơn.')
print('- Crossover_rate cao (1.0) nhanh đạt ngưỡng nhưng dao động; thấp (0.6) ổn định nhưng chậm.')
````

````python
# code vào đây nhé
import numpy as np
import random
import matplotlib.pyplot as plt

# 1. Hàm mục tiêu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# 2. Khởi tạo quần thể (reuse từ lab3: initialize_population nhưng với giới hạn [0,2π])
def initialize_population(pop_size, min_val=0, max_val=2*np.pi):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# 3. Các phương pháp chọn lọc
def select_random(population, fitness_values):
    # Chọn 1 cá thể ngẫu nhiên
    return random.choice(population)

def select_tournament(population, fitness_values, tournament_size=3):
    # reuse từ lab3
    contenders = random.sample(range(len(population)), tournament_size)
    best = max(contenders, key=lambda i: fitness_values[i])
    return population[best]

def select_roulette(population, fitness_values):
    # Chọn theo xác suất tỷ lệ thuận với fitness (dời min thành dương)
    min_f = min(fitness_values)
    shifted = [f - min_f + 1e-6 for f in fitness_values]
    total = sum(shifted)
    probs = [f/total for f in shifted]
    return random.choices(population, weights=probs, k=1)[0]

# 4. Lai ghép & đột biến: reuse from lab3 (arithmetic crossover + mutate real)
def crossover(parent1, parent2, rate=0.8):
    if random.random() < rate:
        return (parent1 + parent2) / 2
    return parent1

def mutate(individual, rate=0.1, min_val=0, max_val=2*np.pi):
    if random.random() < rate:
        return random.uniform(min_val, max_val)
    return individual

# 5. Hàm GA tổng quát với tham số chọn lọc
def genetic_algorithm(selection_fn, pop_size=50, generations=100,
                      crossover_rate=0.8, mutation_rate=0.1):
    pop = initialize_population(pop_size)
    best_history = []
    for gen in range(generations):
        fitness_vals = [fitness_function(x) for x in pop]
        best_history.append(max(fitness_vals))
        new_pop = []
        for _ in range(pop_size):
            p1 = selection_fn(pop, fitness_vals)
            p2 = selection_fn(pop, fitness_vals)
            child = crossover(p1, p2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_pop.append(child)
        pop = new_pop
    return best_history

# 6. Chạy thử & vẽ so sánh
methods = {
    "Random": select_random,
    "Tournament": select_tournament,
    "Roulette": select_roulette
}
plt.figure(figsize=(8,5))
for name, fn in methods.items():
    hist = genetic_algorithm(fn, pop_size=100, generations=100,
                             crossover_rate=0.8, mutation_rate=0.05)
    plt.plot(hist, label=name)
plt.xlabel('Thế hệ')
plt.ylabel('Fitness tốt nhất')
plt.title('So sánh tốc độ hội tụ của các phương pháp chọn lọc')
plt.legend()
plt.show()
````

# Trực quan hóa quần thể – Vẽ biểu đồ phân tán (scatter plot) thể hiện sự phân bố của quần thể qua các thế hệ

---

##  Mục tiêu

- Dùng biểu đồ phân tán (scatter plot) để thể hiện vị trí các cá thể trên không gian biến số qua từng thế hệ.
- Giúp quan sát rõ tiến trình hội tụ, đa dạng di truyền, và sự thay đổi quần thể theo thời gian.

---

## Ý tưởng thực hiện

- Với bài toán tối ưu hàm $k(x, y, z) = x^2 + y^2 + z^2$, không gian tìm kiếm là 3 chiều.
- Mỗi cá thể là một điểm trong không gian $(x, y, z)$.
- Ở mỗi thế hệ, lấy vị trí giải mã của toàn bộ cá thể, vẽ scatter plot.
- Nếu không thể trực quan 3D trực tiếp, có thể vẽ scatter 2D từng cặp biến: $(x, y)$, $(y, z)$, $(x, z)$ qua các thế hệ.

````python
#code vào đây
import numpy as np
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # để vẽ 3D

# --- 1. Hàm mục tiêu 3D ---
def fitness_function_3d(x, y, z):
    # Ở đây k(x,y,z) = x^2 + y^2 + z^2, ta muốn MIN nên đổi dấu để GA MAX
    return -(x**2 + y**2 + z**2)

# --- 2. Khởi tạo quần thể 3D ---
def initialize_population_3d(pop_size, min_val=-5, max_val=5):
    return [
        (random.uniform(min_val, max_val),
         random.uniform(min_val, max_val),
         random.uniform(min_val, max_val))
        for _ in range(pop_size)
    ]

# --- 3. Chọn lọc Tournament (reuse) ---
def select_parents_3d(pop, fitness_vals, tournament_size=3):
    contenders = random.sample(range(len(pop)), tournament_size)
    best_idx = max(contenders, key=lambda i: fitness_vals[i])
    return pop[best_idx]

# --- 4. Lai ghép & đột biến 3D ---
def crossover_3d(p1, p2, rate=0.8):
    if random.random() < rate:
        return (
            (p1[0] + p2[0]) / 2,
            (p1[1] + p2[1]) / 2,
            (p1[2] + p2[2]) / 2
        )
    return p1

def mutate_3d(ind, rate=0.1, min_val=-5, max_val=5):
    x, y, z = ind
    if random.random() < rate:
        x = random.uniform(min_val, max_val)
    if random.random() < rate:
        y = random.uniform(min_val, max_val)
    if random.random() < rate:
        z = random.uniform(min_val, max_val)
    return (x, y, z)

# --- 5. GA 3D, trả về danh sách quần thể qua từng thế hệ ---
def genetic_algorithm_3d(pop_size=100, generations=50,
                         crossover_rate=0.8, mutation_rate=0.1):
    population = initialize_population_3d(pop_size)
    history = [population]  # lưu quần thể thế hệ 0
    for gen in range(generations):
        fitness_vals = [fitness_function_3d(x,y,z) for x,y,z in population]
        new_pop = []
        for _ in range(pop_size):
            p1 = select_parents_3d(population, fitness_vals)
            p2 = select_parents_3d(population, fitness_vals)
            child = crossover_3d(p1, p2, crossover_rate)
            child = mutate_3d(child, mutation_rate)
            new_pop.append(child)
        population = new_pop
        history.append(population)
    return history

# --- 6. Chạy GA và lấy các thế hệ mẫu để vẽ ---
hist = genetic_algorithm_3d(pop_size=200, generations=60,
                            crossover_rate=0.8, mutation_rate=0.05)

# Chọn gen 0, gen giữa và gen cuối để trực quan
gens_to_plot = [0, len(hist)//2, len(hist)-1]
fig = plt.figure(figsize=(12,4))
for i, g in enumerate(gens_to_plot):
    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    pop = hist[g]
    xs, ys, zs = zip(*pop)
    ax.scatter(xs, ys, zs, s=10, alpha=0.6)
    ax.set_title(f'Generation {g}')
    ax.set_xlim(-5,5); ax.set_ylim(-5,5); ax.set_zlim(-5,5)
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('z')

plt.suptitle('Scatter plot quần thể qua các thế hệ')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````
